```python
# Created By: Priyansh Kumar Nigam

# Graphical BlackJack Game

import random
import turtle
drawarea = turtle.Screen()
drawarea.setup(width = 1280, height = 720)
pca = turtle.Turtle()
pca.speed(10)
pca.penup()
pca.goto(-600,120)
pca.pendown()
ace = 0
suits = ['Hearts', 'Clubs', 'Spades', 'Diamonds']
ranks = ['Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace']
value = {'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5, 'Six': 6, 'Seven': 7, 'Eight': 8, 'Nine': 9, 'Ten': 10, 'Jack': 10, 'Queen': 10, 'King': 10, 'Ace': ace}
```

```python
class Card:
    
    def __init__(self, suit, rank):
        
        self.suit = suit
        self.rank = rank
        self.value = value[rank]
        
    def __str__(self):
        
        return f'{self.rank} of {self.suit}'

    def displaycd(self, canum):
        pca.penup()
        if canum == 0:
            pca.forward(300)
        else:
            pca.forward(50)
        pca.pendown()
        for x in range(2):
            a =180
            for i in range(2):
                pca.left(90)
                pca.forward(a)
                a=a/2
        pca.penup()
        pca.left(116.56)
        pca.forward(165)
        if self.value != 10 and self.value != ace and self.value != 0 and self.rank != 'Ace':
            pca.write(self.value, font=("Verdana",15, "normal"))
        elif self.value == 10 and self.rank == 'Ten':
            pca.write(self.value, font=("Verdana",15, "normal"))
        elif self.value == 10 and self.rank == 'Jack':
            pca.write("J", font=("Verdana",15, "normal"))
        elif self.value == 10 and self.rank == 'Queen':
            pca.write("Q", font=("Verdana",15, "normal"))
        elif self.value == 10 and self.rank == 'King':
            pca.write("K", font=("Verdana",15, "normal"))
        elif self.rank == 'Ace' or self.value == ace or self.value != 0:
            pca.write("A", font=("Verdana",15, "normal"))
        pca.backward(65)
        pca.pendown()
        if self.suit == 'Hearts':
            pca.color('red')
            pca.write("♥", font=("Verdana",45, "normal"), align = 'center')
        elif self.suit == 'Clubs':
            pca.color('black')
            pca.write("♣", font=("Verdana",45, "normal"), align = 'center')
        elif self.suit == 'Diamonds':
            pca.color('red')
            pca.write("♦", font=("Verdana",45, "normal"), align = 'center')
        elif self.suit == 'Spades':
            pca.color('black')
            pca.write("♠", font=("Verdana",45, "normal"), align = 'center')
        pca.penup()
        pca.backward(100)
        pca.right(116.56)
        pca.forward(100)
        pca.pendown()
        pca.color('black')
        
        return pca.pos()
```

```python
class Deck:
    
    def __init__(self):
        
        self.all_cards = []
        for suit in suits:
            for rank in ranks:
                createdcard = Card(suit, rank)
                self.all_cards.append(createdcard)
    def shuffle(self):
        
        random.shuffle(self.all_cards)
        
    def deal_one(self):
        
        return self.all_cards.pop(0)
        
    
                
```

```python
class Player:
    
    def __init__(self, name):
        
        self.cards = []
    
    def add_cards(self, new_card):
        self.cards.append(new_card)
        
    def remove1(self):
        return self.cards.pop(0)
    
    def fresh(self):
        
        self.cards = []
```

```python
class bank:
    
    def __init__(self, bal):
        self.bal = bal
        
    def bet(self, val):
        self.val = val
        self.bal = self.bal - self.val
        print('Bet Placed!, balance remaining: {} Rs'.format(self.bal))
    
    def win(self, val):
        self.bal = self.bal + 2*(self.val)
        
    def tie(self, val):
        self.bal = self.bal + self.val
    
    def __str__(self):
        return('Amount Added! Your net balance is now {}'.format(self.bal))
```
